```python
import pandas as pd
import numpy as np
from sklearn import preprocessing
```

Na skupu <code>24-bank.csv</code> izvršiti sledece promene:

a) Učitati skup podataka.

```python
data = pd.read_csv('datasets/24-bank.csv')
```

b) Ispisati prvih nekoliko instanci i dimenzije skupa podataka.

```python
data.head()
```

```python
data.shape
```

c) Prikazati deskriptivne statistike svih atributa.

```python
data.describe()
```

d) Eliminisati instance koje imaju negativnu vrednost atributa <code>SREDNJE_GOD_STANJE_EUR</code>. 

```python
data.replace(data['SREDNJE_GOD_STANJE_EUR'] < 0, np.nan, inplace=True)
```

***II način:***

```python
data['SREDNJE_GOD_STANJE_EUR'] = data['SREDNJE_GOD_STANJE_EUR'].apply(lambda x : np.nan if x < 0 else x)
```

```python
data.isna().any()
```

```python
data.dropna(inplace=True)           #ne moramo navoditi subset=['SREDNJE_GOD_STANJE_EUR'] zato sto 
data.shape                          #je to ionako jedina kolona koja ima nedostajuce vrednosti
```

Uticaj ove transformacije na deskriptivne statistike:

```python
data.describe()
```

e) Vrednosti atributa <code>STAROST</code> podeliti u 5 kategorija jednake širine. 

```python
data['STAROST'] = pd.cut(data['STAROST'], bins=5, labels=[1, 2, 3, 4, 5])
```

```python
data['STAROST'].unique()
```

f) Promeniti kategorije atributa <code>BRACNO_STANJE</code> u <code>u_braku</code> i <code>nije_u_braku</code>.

```python
data['BRACNO_STANJE'].unique()
```

```python
data['BRACNO_STANJE'].replace(['samac', 'razveden(a)'], 'nije_u_braku', inplace=True)
```

```python
data.head()
```

g) Izdvojiti samo numericke atribute.

```python
data.dtypes
```

```python
num_attributes = data.dtypes[(data.dtypes == 'int64') | (data.dtypes == 'float64')].index
num_attributes
```

h) Standardizovati numeričke atribute. Prikazati deskriptivne statistike svih atributa nakon ove transformacije.

```python
scaler = preprocessing.StandardScaler()
```

```python
scaler.fit(data[num_attributes])    #fit - uci raspodele (izracunava mean i std) svakog od atributa
```

```python
data[num_attributes] = scaler.transform(data[num_attributes])
```

Takođe je dostupna bibliotečka podrška i za sledeće tehnike normalizacije:
- Min-max skaliranje - klasa <code>sklearn.preprocessing.MinMaxScaler</code>
- Odsecanje vrednosti - metod <code>pandas.DataFrame.clip</code>
- Logaritamska transformacija - metod <code>pandas.DataFrame.transform('log')</code>
